```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("original.csv")
df = df.iloc[:,1: len(df)]
df.head()
```

```python
X = df.drop('sales',axis=1)
y = df[["sales"]]
```

```python
y
```

```python
X
```

```python
#Statsmodel ile model kurmak...
```

```python
import statsmodels.api as sm
```

```python
lm = sm.OLS(y,X)
```

```python
model = lm.fit()#
```

```python
model.summary()# modelin özetlirini verir
#çoklu dogrusal modelin çıktısı aşağıdadır.
#R^2 bağımsız değişkenlerin bağımlı değişkeni açıklama yüzdesi, Veri setine yeni değerler eklendiğinde bu değer şişer.
#Adj.R^2 düzeltimiş R^2 değeridir. R^2 deki şişmeyi önlemek amacı ile yapılır.
#F-statistic modelin anlamlılığının test etmek için kullanılmaktadır.
#Prob(F-statistic) modelin anlamlığını verir, 0.05'ten küçük olduğu için model anlamlıdır.
#coef kurulacak olan modeldeki bağımsız değişkenlerin katsayılarını temsil etmektedir.
#P>|t| bunun manası modellememk için kullandığımız bğtğn değişkenler anlamldır. Buradaki hipotezler değişkenin anlamlı olmadığını söyleyen hipotezlerdir. Ve bu değerler 0.05'ten küçük olduğu için reddedilir buda bu modeldeki kullanılan değişkenlerin anlamlı oldupu anlamına gelmektedir. 
#Katsayıların yorumlanması: Tv için; diğer değişkenler sabit varsayıldığında TV harcamalarında meydana gelebilecek bir birimlik artış bagımlı değikende 0.0538 birimlik bir artışa sebeb olacaktır denmektedir.
```
